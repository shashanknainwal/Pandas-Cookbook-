```python
# Pandas pivot table replications with groupby aggregation
```

```python
fp= pd.read_csv("flights.csv")
```

```python
fp
```

```python
# Find total number of cancelled flights per origin airport for each airline
```

```python
fp.pivot_table(index="AIRLINE",columns="ORG_AIR",values="CANCELLED",aggfunc='sum',fill_value=0)
```

```python
fg=fp.groupby(["AIRLINE","ORG_AIR"])["CANCELLED"].sum()
fg
```

```python
fg.unstack("ORG_AIR")
```
