###  Grouping and aggregating with multiple columns and functions

1. Find the number of cancelled flights for every airline per weekday 

```python
flights.head()
```

```python
flights.groupby(["AIRLINE","WEEKDAY"])['CANCELLED'].sum()
```

Find the number and percentage of cancelled and diverted flights for every airline per weekday 

```python
flights.groupby(["AIRLINE","WEEKDAY"])['CANCELLED','DIVERTED'].agg(['sum','mean'])
```

For each origin and destination, find the total number of flights, the number and percentage of cancelled flights,the average and varaince of airtime

```python
flights.head(2)
```

```python
flights.groupby(["ORG_AIR",'DEST_AIR'])['CANCELLED','AIR_TIME'].agg(['size','mean','var'])
```

```python
college= pd.read_csv("college.csv")
college
```

```python
college.groupby("STABBR")['UGDS'].agg(['mean','std']).head()
```

![title](2.png) 

```python
def percentagebeteen(s,low,high):
    return s.between(low,high).mean()
    
```

```python
college.groupby(["STABBR",'RELAFFIL'])['UGDS'].agg(percentagebeteen,low=1000,high=10000).head()
```
