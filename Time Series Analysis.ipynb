```python
# Resample on other columns that are not in index
```

```python
crime_new= pd.read_hdf("crime.h5")
```

```python
crime_new
```

```python
monthly_crime=crime_new.resample("M",on='REPORTED_DATE').size()
monthly_crime.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
monthly_crime.plot(title="All Denver Monthly crime")
```

```python
monthly_crime_1=crime_new.resample("W",on='REPORTED_DATE').size()
```

```python
monthly_crime_1.plot(title="All Denver Monthly crime")
```

```python
## Measuring crime by weekday and year
```

```python
crime.resample("W").size()
```

```python
crime_new_crime=crime_new['REPORTED_DATE'].dt.weekday_name.value_counts()
crime_new_crime
# weekends appear to have less crime
```

```python
crime_new_crime.plot(kind='barh')
```

```python
# Crime by year
```

```python
crime_new_crime1=crime_new['REPORTED_DATE'].dt.year.value_counts()
crime_new_crime1
```

```python
crime_new_crime1.plot(kind='barh')
```

```python
# Both together
```

```python
weekday=crime_new['REPORTED_DATE'].dt.weekday_name
year= crime_new['REPORTED_DATE'].dt.year
```

```python
crime_new_final=crime_new.groupby([year,weekday]).size()
```

```python
crime_new_final
```

```python
crime_new_final=crime_new_final.rename_axis(['YEAR','WEEKDAY']).unstack('WEEKDAY')
```

```python
crime_new_final
```

```python
crime_new_final.plot(kind='barh')
```

```python
import seaborn as sns
```

```python
sns.heatmap(crime_new_final)
```
