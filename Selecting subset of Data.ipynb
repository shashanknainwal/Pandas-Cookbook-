** .loc- works with index location
   .iloc- works with integer location

```python
import pandas as pd
```

```python
college= pd.read_csv("college.csv",index_col='INSTNM')
```

```python
college.head()
```

```python
college.iloc[2]
```

```python
college.loc['Amridge University']
```

```python
college.iloc[2:5,1:2]
```

```python
college.iloc[[1,2,5]]
```

```python
college.iloc[[2]]
```

```python
## Selecting both rows and columns from a DF in pandas
```

```python
college.iloc[:1,3:]
```

```python
columns=['CITY', 'STABBR', 'HBCU', 'MENONLY', 'WOMENONLY', 'RELAFFIL',
       'SATVRMID', 'SATMTMID']
```

```python
rows=['Alabama A & M University', 'University of Alabama at Birmingham',
       'Amridge University', 'University of Alabama in Huntsville']
```

```python
college.loc[rows,columns]
```

```python
college.iloc[:10] 
# is same as college.iloc[:10,:] 
```

```python
# Using both index labels and column names
```

```python
start_index=college.index[5]
finish_index= college.index[11]
```

```python
college.loc[start_index:finish_index,['WOMENONLY','SATMTMID']]
```

```python
college[['WOMENONLY','SATMTMID']]
```

```python
college.loc[:'University of Alabama at Birmingham',:]
```
